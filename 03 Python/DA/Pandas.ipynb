To efficiently perform numerical operations, we'll convert relevant parts of the data into NumPy arrays.

### **3.2. Extract Data into Separate Structures**

```python
# Initialize dictionaries to store numerical and categorical data
numerical_data = {col: [] for col in numerical_cols}
categorical_data = {col: [] for col in categorical_cols}
income_data = []

# Populate the data
for row in df:
    # Numerical data
    for col in numerical_cols:
        value = row[col].strip()
        if value.isdigit():
            numerical_data[col].append(int(value))
        else:
            numerical_data[col].append(np.nan)  # Handle non-integer values as NaN
    
    # Categorical data
    for col in categorical_cols:
        value = row[col].strip()
        categorical_data[col].append(value if value else None)  # Replace empty strings with None
    
    # Target variable
    income_data.append(row['income'].strip())

# Convert to NumPy arrays
num_arrays = {col: np.array(vals, dtype=float) for col, vals in numerical_data.items()}
cat_arrays = {col: np.array(vals) for col, vals in categorical_data.items()}
income_array = np.array(income_data)

print("Data conversion to NumPy arrays completed.")
```

**Explanation:**

- **numerical_data**: Stores numerical columns as lists.
- **categorical_data**: Stores categorical columns as lists.
- **Handling Missing Values**: Non-integer numerical entries are converted to `NaN`, and empty categorical entries are set to `None`.
- **Conversion**: Lists are converted into NumPy arrays for efficient computation.

## **4. Handling Missing Values**

We'll identify and handle missing values in both numerical and categorical data.

### **4.1. Identify Missing Values**

```python
# Function to count missing values in numerical arrays
def count_missing_numerical(arr):
    return np.isnan(arr).sum()

# Function to count missing values in categorical arrays
def count_missing_categorical(arr):
    return np.sum(arr == None)

# Count missing values in numerical columns
print("\nMissing Values in Numerical Columns:")
for col, arr in num_arrays.items():
    missing = count_missing_numerical(arr)
    print(f"{col}: {int(missing)} missing values")

# Count missing values in categorical columns
print("\nMissing Values in Categorical Columns:")
for col, arr in cat_arrays.items():
    missing = count_missing_categorical(arr)
    print(f"{col}: {int(missing)} missing values")
```

### **4.2. Handle Missing Values**

For simplicity, we'll remove rows with any missing values. Alternatively, you can choose to impute missing values based on your requirements.

```python
# Identify indices with no missing values
valid_indices = np.ones(len(num_arrays['age']), dtype=bool)

# Check numerical columns for NaN
for col in numerical_cols:
    valid_indices &= ~np.isnan(num_arrays[col])

# Check categorical columns for None
for col in categorical_cols:
    valid_indices &= (cat_arrays[col] != None)

# Count valid rows before filtering
initial_count = len(income_array)
print(f"\nInitial number of rows: {initial_count}")

# Filter data based on valid_indices
for col in numerical_cols:
    num_arrays[col] = num_arrays[col][valid_indices]
for col in categorical_cols:
    cat_arrays[col] = cat_arrays[col][valid_indices]
income_array = income_array[valid_indices]

# Count valid rows after filtering
final_count = len(income_array)
print(f"Number of rows after removing missing values: {final_count}")
print(f"Number of rows removed: {initial_count - final_count}")
```

**Sample Output:**
```

Number of rows after removing missing values: 28471
Number of rows removed: 4090
```

**Explanation:**

- **valid_indices**: A boolean array where `True` indicates rows without any missing values.
- **Filtering**: Retains only the rows without missing values across all columns.

## **6. Analyzing Distributions**

We'll explore the distribution of numerical features and the frequency of categorical features.

### **6.1. Numerical Features Distribution**

We'll create simple textual histograms for numerical features.

```python
def textual_histogram(arr, bins=10):
    hist, bin_edges = np.histogram(arr, bins=bins)
    for i in range(len(hist)):
        range_start = bin_edges[i]
        range_end = bin_edges[i+1]
        count = hist[i]
        bar = '*' * (count // (max(hist) // 50 + 1))  # Scale bars to a maximum length
        print(f"{range_start:.1f} - {range_end:.1f}: {count} | {bar}")

print("\nNumerical Features Distribution:")

for col in numerical_cols:
    arr = num_arrays[col]
    print(f"\nColumn: {col}")
    textual_histogram(arr, bins=10)
```

**Sample Output:**
```
Numerical Features Distribution:

Column: age
17.0 - 25.3: 2000 | *********************
25.3 - 33.6: 5000 | ****************************************
...
```

**Explanation:**

- **textual_histogram**: Generates a simple textual histogram by scaling the number of asterisks based on the count in each bin.
- **Scaling**: Adjusts the length of bars to prevent excessively long lines.

### **7.2. Categorical Features vs. Income**

We'll calculate the frequency and percentage distribution of each category within the target variable.

```python
print("\nCategorical Features Distribution by Income:")

for col in categorical_cols:
    print(f"\nFeature: {col}")
    categories = np.unique(cat_arrays[col])
    for category in categories:
        mask = (cat_arrays[col] == category)
        income_counts = Counter(income_array[mask])
        total = sum(income_counts.values())
        if total == 0:
            continue  # Avoid division by zero
        print(f"  Category: {category}")
        for income, count in income_counts.items():
            percentage = (count / total) * 100
            print(f"    {income}: {count} ({percentage:.2f}%)")
```

**Sample Output:**

```
    >50K: 1250 (25.00%)
  ...
```

**Interpretation:**

- **Workclass**: A significant portion of individuals in the `Private` and `Self-emp-not-inc` categories earn `<=50K`, while a smaller segment earns `>50K`.
- Similar interpretations can be made for other categorical features.
