# Plotting Numerical and Categorical Data

Understanding the nature of your data—whether numerical or categorical—is crucial in selecting the appropriate visualization. This section explores various plot types tailored for numerical and categorical datasets.

### Numerical Data Plots

Numerical data represents quantities and can be continuous or discrete. Common visualizations include line plots, scatter plots, histograms, and box plots.

#### Line Plot (`plot`)

**Function:** `plt.plot()`

**Description:** Creates a 2D line plot, connecting data points with lines.

**Parameters:**

- **`x`** *(array-like)*: X-coordinates of data points.
- **`y`** *(array-like)*: Y-coordinates of data points.
- **`fmt`** *(str, optional)*: Format string specifying line style and color (e.g., `'ro-'` for red circles with solid lines).
- **`label`** *(str, optional)*: Label for the plot (used in legends).
- **`color`** *(color, optional)*: Line color.
- **`linestyle`** *(str, optional)*: Line style (`'-'`, `'--'`, `':'`, `'-.`).
- **`marker`** *(str, optional)*: Marker style (`'o'`, `'s'`, `'^'`, etc.).
- **`linewidth`** or **`lw`** *(float, optional)*: Width of the line.
- **`markersize`** or **`ms`** *(float, optional)*: Size of the markers.

**Example:**

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a line plot
plt.plot(x, y, color='blue', linestyle='-', marker='o', label='Sine Wave')

# Add title and labels
plt.title('Sine Wave Line Plot')
plt.xlabel('X-axis')
plt.ylabel('sin(X)')

# Show legend
plt.legend()

# Display grid
plt.grid(True)

# Show the plot
plt.show()
```

**Output:**

A sine wave plotted with blue solid lines and circular markers, accompanied by a legend and gridlines.

#### Scatter Plot (`scatter`)

**Function:** `plt.scatter()`

**Description:** Creates a scatter plot of y vs. x with varying marker sizes and colors.

**Parameters:**

- **`x`** *(array-like)*: X-coordinates of data points.
- **`y`** *(array-like)*: Y-coordinates of data points.
- **`s`** *(float or array-like, optional)*: Size of markers.
- **`c`** *(color or array-like, optional)*: Color of markers.
- **`marker`** *(str, optional)*: Marker style.
- **`cmap`** *(colormap, optional)*: Colormap for marker colors.
- **`alpha`** *(float, optional)*: Transparency level (0.0 to 1.0).
- **`label`** *(str, optional)*: Label for the plot.

**Example:**

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
np.random.seed(0)
x = np.random.randn(50)
y = np.random.randn(50)
sizes = 1000 * np.random.rand(50)
colors = np.random.rand(50)

# Create a scatter plot
plt.scatter(x, y, s=sizes, c=colors, alpha=.2, cmap='Wistia', marker='o', label='Random Points')

# Add title and labels
plt.title('Random Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Show colorbar
plt.colorbar(label='Color Scale')

# Show legend
plt.legend()

# Display grid
plt.grid(True)

# Show the plot
plt.show()
```

**Output:**

A scatter plot with triangular markers of varying sizes and colors, a colorbar indicating the color scale, legend, and gridlines.

#### Histogram (`hist`)

**Function:** `plt.hist()`

**Description:** Plots the distribution of numerical data by grouping data into bins and counting the number of observations in each bin.

**Parameters:**

- **`x`** *(array-like)*: Input data.
- **`bins`** *(int or sequence, optional)*: Number of bins or bin edges.
- **`range`** *(tuple, optional)*: Lower and upper range of the bins.
- **`density`** *(bool, optional)*: If `True`, the histogram is normalized.
- **`color`** *(color, optional)*: Color of the bars.
- **`edgecolor`** *(color, optional)*: Color of the bar edges.
- **`alpha`** *(float, optional)*: Transparency level.
- **`label`** *(str, optional)*: Label for the histogram.

**Example:**

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = np.random.normal(loc=0, scale=1, size=1000)

# Create a histogram
plt.hist(data, bins=30, density=False, color='skyblue', edgecolor='black', alpha=0.7, label='Normal Distribution')

# Add title and labels
plt.title('Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show legend
plt.legend()

# Display grid
plt.grid(True)

# Show the plot
plt.show()
```

**Output:**

A histogram displaying the distribution of normally distributed data with sky blue bars, black edges, transparency, legend, and gridlines.

#### Box Plot (`boxplot`)

**Function:** `plt.boxplot()`

**Description:** Creates a box plot to visualize the distribution of numerical data, highlighting the median, quartiles, and potential outliers.

**Parameters:**

- **`x`** *(array-like or sequence of arrays)*: Input data.
- **`notch`** *(bool, optional)*: If `True`, draws a notched box plot.
- **`vert`** *(bool, optional)*: If `True`, boxes are vertical.
- **`patch_artist`** *(bool, optional)*: If `True`, fills boxes with color.
- **`labels`** *(sequence, optional)*: Labels for each dataset.
- **`showmeans`** *(bool, optional)*: If `True`, shows the mean as a marker.
- **`meanline`** *(bool, optional)*: If `True`, shows the mean as a line.
- **`showfliers`** *(bool, optional)*: If `True`, displays outliers.
- **`boxprops`**, **`medianprops`**, **`meanprops`**, etc. *(dict, optional)*: Properties for box elements.

**Example:**

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
np.random.seed(0)
data1 = np.random.normal(0, 1, 100)             # 99% in (-3,3)
data2 = np.random.normal(5, 2, 100)             # 99% in (-1,11)
data3 = np.random.normal(-3, 1.5, 100)          # 99% in (-7.5,1.5)

# Create a box plot
plt.boxplot([data1, data2, data3], notch=True, vert=True, patch_artist=True,
            labels=['Data 1', 'Data 2', 'Data 3'],
            showmeans=True,
            boxprops=dict(facecolor='lightgreen', color='green'),
            medianprops=dict(color='red'),
            meanprops=dict(marker='D', markeredgecolor='black', markerfacecolor='yellow'))

# Add title and labels
plt.title('Box Plot of Three Datasets')
plt.ylabel('Value')

# Show grid
plt.grid(True)

# Show the plot
plt.show()
```

**Output:**

A notched box plot displaying three datasets with customized colors, mean markers, median lines, labels, and gridlines.

### Categorical Data Plots

Categorical data represents distinct groups or categories. Common visualizations include bar charts, horizontal bar charts, and pie charts.

#### Bar Chart (`bar`)

**Function:** `plt.bar()`

**Description:** Creates a vertical bar chart for comparing categorical data.

**Parameters:**

- **`x`** *(sequence)*: Categories or positions of the bars.
- **`height`** *(sequence)*: Heights of the bars.
- **`width`** *(float, optional)*: Width of the bars.
- **`bottom`** *(float or array-like, optional)*: Starting position for each bar.
- **`align`** *(str, optional)*: Alignment of the bars (`'center'`, `'edge'`).
- **`color`** *(color or sequence, optional)*: Color(s) of the bars.
- **`edgecolor`** *(color or sequence, optional)*: Color(s) of the bar edges.
- **`alpha`** *(float, optional)*: Transparency level.
- **`label`** *(str, optional)*: Label for the bars.

**Example:**

```python
import matplotlib.pyplot as plt

# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [25, 40, 15, 30]

# Create a bar chart
plt.bar(categories, values, color='coral', edgecolor='black', alpha=0.7, label='Sales')

# Add title and labels
plt.title('Sales by Category')
plt.xlabel('Categories')
plt.ylabel('Sales Units')

# Show legend
plt.legend()

# Display grid
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

**Output:**

A vertical bar chart comparing sales units across four categories with coral-colored bars, black edges, legend, and horizontal gridlines.

#### Horizontal Bar Chart (`barh`)

**Function:** `plt.barh()`

**Description:** Creates a horizontal bar chart for comparing categorical data.

**Parameters:**

- **`y`** *(sequence)*: Categories or positions of the bars.
- **`width`** *(sequence)*: Widths of the bars.
- **`height`** *(float, optional)*: Height of the bars.
- **`left`** *(float or array-like, optional)*: Starting position for each bar.
- **`align`** *(str, optional)*: Alignment of the bars (`'center'`, `'edge'`).
- **`color`** *(color or sequence, optional)*: Color(s) of the bars.
- **`edgecolor`** *(color or sequence, optional)*: Color(s) of the bar edges.
- **`alpha`** *(float, optional)*: Transparency level.
- **`label`** *(str, optional)*: Label for the bars.

**Example:**

```python
import matplotlib.pyplot as plt

# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [25, 40, 15, 30]

# Create a horizontal bar chart
plt.barh(categories, values, color='skyblue', edgecolor='black', alpha=0.7, label='Sales')

# Add title and labels
plt.title('Sales by Category (Horizontal)')
plt.xlabel('Sales Units')
plt.ylabel('Categories')

# Show legend
plt.legend()

# Display grid
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

**Output:**

A horizontal bar chart comparing sales units across four categories with sky blue bars, black edges, legend, and vertical gridlines.

#### Pie Chart (`pie`)

**Function:** `plt.pie()`

**Description:** Creates a pie chart to represent proportions of a whole.

**Parameters:**

- **`x`** *(array-like)*: Proportions or sizes of the pie slices.
- **`labels`** *(sequence, optional)*: Labels for each slice.
- **`colors`** *(sequence, optional)*: Colors for the slices.
- **`autopct`** *(str or function, optional)*: String or function to label slices with their numeric value.
- **`startangle`** *(float, optional)*: Starting angle of the pie chart.
- **`explode`** *(sequence, optional)*: Fraction of radius to offset each slice.
- **`shadow`** *(bool, optional)*: Draw a shadow.
- **`radius`** *(float, optional)*: Radius of the pie chart.
- **`counterclock`** *(bool, optional)*: If `True`, draws the pie chart counter-clockwise.

**Example:**

```python
import matplotlib.pyplot as plt

# Sample data
labels = ['Category A', 'Category B', 'Category C', 'Category D']
sizes = [30, 45, 15, 10]
colors = ['gold', 'lightcoral', 'lightskyblue', 'yellowgreen']
explode = (0.1, 0, 0, 0)  # Explode the first slice

# Create a pie chart
plt.pie(sizes, labels=labels, colors=colors, explode=explode,
        autopct='%1.1f%%', shadow=True, startangle=140)

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Add title
plt.title('Sales Distribution by Category')

# Show the plot
plt.show()
```

**Output:**

A pie chart showing the sales distribution across four categories with the first slice exploded, percentage labels, shadow effect, and custom colors.

## Single Variable Analysis

Single Variable Analysis focuses on exploring and understanding the characteristics of a single variable. It involves summarizing the data and visualizing it to understand its distribution, central tendency, dispersion, and outliers.

---

### Steps for Single Variable Analysis

1. **Load the Data**  
   Ensure the data is properly imported and cleaned before proceeding.

2. **Understand the Variable**  
   Identify the variable type:
   - **Numerical**: Continuous (e.g., height, weight) or Discrete (e.g., number of items).
   - **Categorical**: Ordinal (e.g., rankings) or Nominal (e.g., gender, city).

3. **Summary Statistics**  
   Use descriptive statistics to summarize the data:
   - **For Numerical Variables**:
     - Mean, Median, Mode
     - Minimum, Maximum
     - Standard Deviation, Variance
     - Quartiles (Q1, Q3), Interquartile Range (IQR)
   - **For Categorical Variables**:
     - Frequency distribution
     - Percentage distribution

4. **Visualize the Data**  
   Visualization tools help in identifying patterns, trends, and anomalies:
   - **For Numerical Variables**:
     - Histogram
     - Box Plot
     - Density Plot
   - **For Categorical Variables**:
     - Bar Plot
     - Pie Chart

5. **Check for Outliers**  
   For numerical variables, use:
   - Box Plot
   - Z-Score
   - IQR method (\( Q3 + 1.5 \times IQR \), \( Q1 - 1.5 \times IQR \))

6. **Interpret the Results**  
   Derive insights from the analysis to inform further exploration or decision-making.

---

### Example: Single Variable Analysis on a Numerical Variable

#### **Dataset**: `data1` (from your earlier example)

1. **Summary Statistics**:
   - Mean, Median, Mode
   - Standard Deviation
   - Min, Max, IQR

2. **Visualizations**:
   - Histogram to check the distribution.
   - Box Plot to detect outliers.

#### **Code Example** (for `data1`):

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Data generation
np.random.seed(0)
data1 = np.random.normal(12, 3, 100)

# Summary statistics
mean = np.mean(data1)
median = np.median(data1)
std_dev = np.std(data1)
min_val = np.min(data1)
max_val = np.max(data1)
q1 = np.percentile(data1, 25)
q3 = np.percentile(data1, 75)
iqr = q3 - q1

print(f"Mean: {mean:.2f}, Median: {median:.2f}, Std Dev: {std_dev:.2f}")
print(f"Min: {min_val:.2f}, Max: {max_val:.2f}")
print(f"Q1: {q1:.2f}, Q3: {q3:.2f}, IQR: {iqr:.2f}")

# Visualizations
plt.figure(figsize=(12, 5))

# Histogram
plt.subplot(1, 2, 1)
plt.hist(data1, bins=10, edgecolor='black')
plt.title('Histogram of data1')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Box Plot
plt.subplot(1, 2, 2)
sns.boxplot(data1)
plt.title('Box Plot of data1')

plt.tight_layout()
plt.show()
```

Let's use a real dataset for single-variable analysis. For this purpose, we can use the **Iris dataset**, a well-known dataset in machine learning and data analysis. The Iris dataset contains measurements of flower attributes (sepal length, sepal width, petal length, and petal width) for three species of Iris flowers.


### **Dataset**: Iris Dataset  
The dataset contains 150 records with the following attributes:
- **Sepal Length (cm)**
- **Sepal Width (cm)**
- **Petal Length (cm)**
- **Petal Width (cm)**
- **Species** (Setosa, Versicolor, Virginica)

### Single Variable Analysis Example
Let’s analyze the **Sepal Length** attribute.

#### Steps:
1. **Load the Dataset**:
   Use `pandas` to load the dataset.
2. **Summary Statistics**:
   Calculate mean, median, standard deviation, and quartiles for Sepal Length.
3. **Visualizations**:
   - Histogram
   - Box Plot


### Code Implementation

```python
# Import necessary libraries
import pandas as pd

# Fetch the Iris dataset from UCI Repository
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
columns = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth', 'Species']
iris = pd.read_csv(url, header=None, names=columns)

# Basic dataset overview
iris.head(), iris.info()
```

```python
import matplotlib.pyplot as plt

# Pairplot-like exploration
plt.figure(figsize=(16, 12))
features = ['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth']

# Enhancing the visualization with mean lines on histograms

plt.figure(figsize=(16, 12))

for i, column in enumerate(features, 1):
    data = iris[column]

    # Histogram with mean line
    plt.subplot(4, 2, 2 * i - 1)
    plt.hist(data, bins=10, color='skyblue', edgecolor='black', alpha=0.7)
    plt.axvline(data.mean(), color='red', linestyle='dashed', linewidth=2, label=f'Mean: {data.mean():.2f}')
    plt.title(f'Histogram of {column}')
    plt.xlabel(f'{column} (cm)')
    plt.ylabel('Frequency')
    plt.legend()

    # Box Plot
    plt.subplot(4, 2, 2 * i)
    plt.boxplot(data, vert=False, patch_artist=True, boxprops=dict(facecolor='lightgreen'))
    plt.title(f'Box Plot of {column}')
    plt.xlabel(f'{column} (cm)')

plt.tight_layout()
plt.show()
```

```python
from IPython.display import display


summary_stats = iris[features].describe().T

# Using IPython to display the summary statistics
display(summary_stats.style.set_caption("Summary Statistics for Numerical Variables (Mock Iris Dataset)")
        .format("{:.2f}")
        .background_gradient(cmap='coolwarm', subset=['mean', 'std']))
```

### Correlation analysis and heatmap visualization

```python
# Calculating the correlation matrix
correlation_matrix = iris[features].corr()

# Plotting the heatmap
plt.figure(figsize=(8, 6))
plt.imshow(correlation_matrix, cmap='Blues', interpolation='nearest')
plt.colorbar(label='Correlation Coefficient')
plt.title('Feature Correlation Heatmap', fontsize=14)
plt.xticks(range(len(features)), features, rotation=45)
plt.yticks(range(len(features)), features)
plt.clim(-1, 1)  # Ensuring the color range is between -1 and 1

# Adding correlation coefficients to the heatmap
for i in range(len(features)):
    for j in range(len(features)):
        plt.text(j, i, f"{correlation_matrix.iloc[i, j]:.2f}",
                 ha='center', va='center', color='white' if abs(correlation_matrix.iloc[i, j]) > 0.5 else 'black')

plt.tight_layout()
plt.show()
```
