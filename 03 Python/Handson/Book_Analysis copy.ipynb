This dataset contains information about bestselling books, with the following columns:

1. **Name**: The title of the book.
2. **Author**: The author of the book.
3. **User Rating**: Average rating from users (out of 5).
4. **Reviews**: Number of user reviews.
5. **Price**: Price of the book.
6. **Year**: Year the book was a bestseller.
7. **Genre**: Fiction or Non-Fiction category.
