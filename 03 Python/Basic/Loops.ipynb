# 🔄 **Python Loops – Comprehensive Tutorial**

Loops in Python are used to **repeat a block of code** multiple times, making it easier to automate repetitive tasks. In this tutorial, we’ll cover the types of loops available in Python, their syntax, use cases, and provide extensive examples.

---

## 📌 **Types of Loops in Python**

Python provides two main types of loops:

1. **`for` Loop**: Iterates over a sequence (e.g., list, tuple, string).
2. **`while` Loop**: Repeats as long as a condition is `True`.

---

## 🔹 **1. The `for` Loop**

The **`for` loop** is used to iterate over a sequence (e.g., list, tuple, string, or range). For each iteration, the loop variable takes on the value of the next element in the sequence.

### **Syntax**

```python
for variable in sequence:
    # Code to execute for each element in the sequence
```

### **Examples**

#### **Example 1: Loop Through a List**

```python
fruits = ["apple", "banana", "cherry"]

for fruit in fruits:
    print(fruit)
```

#### **Output**
```
apple
banana
cherry
```

#### **Example 2: Loop Through a String**

```python
word = "Python"

for letter in word:
    print(letter)
```

#### **Output**
```
P
y
t
h
o
n
```

### **Using the `range()` Function**

The `range()` function generates a sequence of numbers. It is commonly used with `for` loops.

#### **Syntax of `range()`**

- **`range(stop)`** – Generates numbers from `0` to `stop - 1`.
- **`range(start, stop)`** – Generates numbers from `start` to `stop - 1`.
- **`range(start, stop, step)`** – Generates numbers from `start` to `stop - 1`, incrementing by `step`.

#### **Example 3: Loop Using `range()`**

```python
# Loop from 0 to 4
for i in range(5):
    print(i)
```

#### **Output**
```
0
1
2
3
4
```

#### **Example 4: Loop with Start and Stop**

```python
# Loop from 2 to 5
for i in range(2, 6):
    print(i)
```

#### **Output**
```
2
3
4
5
```

#### **Example 5: Loop with Step Size**

```python
# Loop from 1 to 9 with a step of 2
for i in range(1, 10, 2):
    print(i)
```

#### **Output**
```
1
3
5
7
9
```

---

## 🔹 **2. The `while` Loop**

The **`while` loop** repeats a block of code as long as a specified condition is `True`.

### **Syntax**

```python
while condition:
    # Code to execute while the condition is True
```

### **Examples**

#### **Example 1: Simple `while` Loop**

```python
count = 0

while count < 5:
    print(count)
    count += 1
```

#### **Output**
```
0
1
2
3
4
```

#### **Example 2: Infinite Loop with Break Condition**

```python
count = 0

while True:
    print(count)
    count += 1
    if count == 3:
        break
```

#### **Output**
```
0
1
2
```

---

## 🔹 **Loop Control Statements**

Python provides control statements to manage the flow of loops:

1. **`break`**: Exits the loop immediately.
2. **`continue`**: Skips the current iteration and continues with the next iteration.
3. **`else`**: Executes a block of code when the loop finishes normally (without a `break`).

### **1. The `break` Statement**

Exits the loop when a specific condition is met.

```python
for i in range(5):
    if i == 3:
        break
    print(i)
```

#### **Output**
```
0
1
2
```

### **2. The `continue` Statement**

Skips the current iteration and continues with the next one.

```python
for i in range(5):
    if i == 2:
        continue
    print(i)
```

#### **Output**
```
0
1
3
4
```

### **3. The `else` Clause with Loops**

The `else` block runs when the loop finishes without a `break`.

```python
for i in range(3):
    print(i)
else:
    print("Loop finished successfully.")
```

#### **Output**
```
0
1
2
Loop finished successfully.
```

---

## 🔹 **Nested Loops**

You can place one loop inside another loop. This is useful for working with multi-dimensional data.

### **Example: Nested `for` Loops**

```python
for i in range(3):
    for j in range(2):
        print(f"i = {i}, j = {j}")
```

#### **Output**
```
i = 0, j = 0
i = 0, j = 1
i = 1, j = 0
i = 1, j = 1
i = 2, j = 0
i = 2, j = 1
```

## 📌 **Common Mistakes with Loops**

1. **Infinite Loop**: Forgetting to update the loop variable in a `while` loop.

    ```python
    # Infinite loop example (don't run this!)
    count = 0
    while count < 5:
        print(count)
    ```

2. **Incorrect Indentation**: Ensure your code blocks are properly indented.

    ```python
    # Correct
    for i in range(3):
        print(i)

    # Incorrect
    for i in range(3):
    print(i)  # This will raise an IndentationError
    ```
